# Arithmetic Functions

```python
print("-------------A--------------")
A = np.arange(10)
print(A)
print(A.shape)
print("-------------A after adding axis--------------")
C = A[np.newaxis,:]
print(C)
print(C.shape)
print("-------------B--------------")
B = np.random.randint(0,10, (1,10))
print(B)
print(B.shape)
```

#### Add Arrays

```python
D = A + C
print(D)
print(D.shape)
E = np.add(A, C)
print(E)
print(E.shape)
```

#### Squeeze additional dimensions

```python
E = D.squeeze()
print(E)
print(E.shape)
```

#### Sum all the values

```python
from numpy._core.multiarray import arange
a_sum = np.sum(D)
print(a_sum)
print("-------------In below, axis-0 is 2, axis-1 is 2, axis-2 is 5 and axis-3 is 4--------------")
n_array = np.arange(80).reshape(2, 2, 5, 4)
print(n_array)
print(n_array.shape)
print("-----------Sum by axis = 0 ------")
sum_array = n_array.sum(axis=0)
print(sum_array)
print(sum_array.shape)
```
